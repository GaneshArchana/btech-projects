```python
import pandas as pd 
import numpy as np
from scipy.optimize import minimize 
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set()
```

```python
df= pd.read_csv("STOCK20.csv")
df.set_index("Date",inplace=True)
return_df=df.pct_change(1).dropna()
return_df.tail()
```

```python
def getportreturn(weigths):
    exp_ret_portfolio=np.dot(np.transpose(weigths),return_df.mean())*250
    return exp_ret_portfolio   
```

```python
return_df
```

```python
num_stocks=len(return_df.columns)
init_weigths=[1/num_stocks]*num_stocks
init_weigths
```

```python
getportreturn(init_weigths)
```

```python
target_return_peter=0.20
target_return_patrik=0.14
```

72/return_ann=no of years to double the money

```python
bounds = tuple((0,1) for i in range(num_stocks))
bounds
```

```python
cons1 = ({'type': 'eq', 'fun':lambda w: np.sum(w)-1},
         {'type':'eq','fun':lambda x : x.dot(return_df.mean())*250-target_return_peter})
cons2 = ({'type': 'eq', 'fun':lambda w: np.sum(w)-1},
         {'type':'eq','fun':lambda x : x.dot(return_df.mean())*250-target_return_patrik})
```

```python
results1 = minimize(fun=getportreturn,x0=init_weigths,bounds=bounds, constraints=cons1)
results1
```

```python
results2 = minimize(fun=getportreturn,x0=init_weigths,bounds=bounds, constraints=cons2)
results2
```
